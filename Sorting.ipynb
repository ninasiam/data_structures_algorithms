*Bubble sort*

1.   The simplest sorting algorithm
2.   Compare each element with the next one
3.   O(n^2) time complexity, O(1) space complexity



```
# Bubble Complexity: O(n^2)
def bubble_sort(input_list):
  # length of input list
  size_l = len(input_list)
  target_num_sorted_pairs = size_l - 1                                          # the number of comparisons at each round
  num_sorted_pairs = 0
  i = 0
  while num_sorted_pairs != target_num_sorted_pairs:                            # when all the pairs in a round are sorted -> exit
    num_sorted_pairs = 0                                                        # at each round we set to zero the sorted pairs 
    i = 0
    while i < size_l - 1:                                                       
      j = i + 1
      if input_list[i] > input_list[j]:
        # swap the values
        input_list[j], input_list[i] = input_list[i], input_list[j]
      else:
        num_sorted_pairs += 1                                                   # mark the sorted pairs
      i += 1
  return input_list

def bubble_sort2(input_list):
  # length of input list
  size_l = len(input_list)
  for comps in range(size_l):                                                   # number of rounds                  
    for j in range(size_l-1):                                                       
      if input_list[j] > input_list[j + 1]:
        # swap the values
        input_list[j+1], input_list[j] = input_list[j], input_list[j+1]
  return input_list
```

*Selection Sort*

1.   One of the simplest algorithms
2.   O(n^2) time complexity, O(1) space complexity
3.   We keep track of the smallest item for n (size of list) rounds,




```
def selection_sort(input_list):
  # length of input list
  size_l = len(input_list)
  current_smallest_val = 0                                                      # position of the smallest index
  for i in range(size_l):                                                       # for each index of the list
    current_smallest_val = i                                                                     
    for smaller_val_ahead in range(i+1, size_l):                                                        
      if input_list[smaller_val_ahead] < input_list[current_smallest_val]:
        current_smallest_val = smaller_val_ahead
    # swap the values (the current index with the smallest value we found in this search round)
    input_list[current_smallest_val], input_list[i] = input_list[i], input_list[current_smallest_val]
  return input_list
```

*Insertion Sort*

1.   Not very efficient, **UNLESS** the list is almost sorted (or sorted)!
2.   In this case we can have O(n)
3.   In small datasets, its very efficient



```
def insertion_sort(input_list):
  # length of input list
  size_l = len(input_list)
  for j in range(size_l - 1):
    if input_list[j + 1] < input_list[j]:
      # swap items
      input_list[j], input_list[j + 1] = input_list[j + 1], input_list[j]
      # check all the previous
      for k in range(j, 0,-1):
        print(k)
        if input_list[k] < input_list[k - 1]:
          input_list[k], input_list[k - 1] = input_list[k - 1], input_list[k]
        else:
          break
    print(input_list)

  return input_list
```

*Merge Sort*

1.   Divide and Conquer as Quick sort 
2.   Recursive method
3.   Complexity of O(nlogn)
4.   Typically performs better than any other sort
5.   But has space complexity -> O(n)


```
def merge_sort(input_list):

  # function that merges and compares the unit elements
  def merge(sub_left, sub_right):
    i = j = 0              
    print("Mpike")                                                              # indices for the left sublist right sublist 
    sub_list_sorted = []
    while i < len(sub_left) and j < len(sub_right):                             # we have elements to compare
      if sub_left[i] < sub_right[j]:
        # the element of the left is smaller
        sub_list_sorted.append(sub_left[i])
        i += 1 # increase the index on the left list
      else:
        # the element of the right is smaller
        sub_list_sorted.append(sub_right[j])
        j += 1 # increase the index on the right list

    # append the elements of left list since the are no elements on the right
    while i < len(sub_left):
      sub_list_sorted.append(sub_left[i])
      i += 1

    # append the elements of left list since the are no elements on the left
    while j < len(sub_right):
      sub_list_sorted.append(sub_right[j])
      j += 1

    # return the soted sublist
    return sub_list_sorted

  # stopping criterion
  if len(input_list) == 1:
    return input_list
  
  # split the array to left and right
  left  = input_list[:len(input_list) // 2]
  right = input_list[len(input_list) // 2:]
  print(left)
  print(right)

  # recursivelly break down left sublist
  return merge(merge_sort(left), merge_sort(right))
```
