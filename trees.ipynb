**Binary Search Tree**
*   Very good at search!
*   The allow as to preserve relationships
*   Ordered
*   Flexible Size
*   No one O(1) operations



> All the nodes at right of a node are larger, on the left are smaller
 Very fast lookup in contrast to a linear data structure. However, insert and delete are O(logn). The right branch is strictly increasing, the left branch is strictly decreasing!

```
class Node:
  def __init__(self, value):
    self.value = value
    self.right = None
    self.left = None

class BinarySearchTree:
  def __init__(self):
    self.root = None

  def __str__(self):
    return str(self.__dict__)

  def insert(self, value):
    # insert an element to the tree
    found_position = False
    new_node = Node(value)
    if self.root:
      current_node = self.root
      while not found_position:
        if value > current_node.value:
          # go right
          if current_node.right:
            current_node = current_node.right
          else:
            current_node.right = new_node
            found_position = True
        else:
          # go left
          if current_node.left:
            print("MPIKE2")
            current_node = current_node.left
          else:
            print("MPIKE")
            current_node.left = new_node
            found_position = True
    else:
      print("MPIKE")
      self.root = new_node
          

  def lookup(self, value):
    # lookup an element to the tree
    found_element = False
    if self.root:
      current_node = self.root
      while not found_element and current_node:
        if value == current_node.value:
          found_element = True
          print(f"Found an element with value {value}")
        else:
          if value > current_node.value:
            # go right
            current_node =current_node.right
          else:
            # go left
            current_node = current_node.left
   
  def remove(self, value):
    # remove an item
    found_element = False
    right = False
    left = False
    if self.root:
      current_node = self.root
      parent_node = None
      while not found_element and current_node:
        if value > current_node.value:
          # go right
          parent_node = current_node
          current_node = current_node.right
          right = True
        elif value < current_node.value:
          # go left
          parent_node = current_node
          current_node = current_node.left
          left = True
        else: 
          if current_node.left == None and current_node.right == None: # leaf case
            # we found a leaf
            if parent_node == None:
              current_node = None
              found_element = True
            elif parent_node.value > current_node.value:
              parent_node.left = None
              found_element = True
            else:
              parent_node.right = None
              found_element = True

          elif current_node.left == None and current_node.right != None: # null left child
            if parent_node == None:
              self.root = current_node.right
              return
            else:
              if parent_node.value > current_node.value:
                parent_node.left = current_node.right
              else:
                parent_node.right = current_node.right
            found_element = True

          elif current_node.left != None and current_node.right == None: # null right child
            if parent_node == None:
              self.root = current_node.left
              return
            else:
              if parent_node.value > current_node.value:
                parent_node.left = current_node.left
              else:
                parent_node.right = current_node.left
            found_element = True
          else:
            del_node = current_node.right
            del_node_parent = current_node.right
            while del_node.left:
              del_node_parent = del_node
              del_node = del_node.left
            current_node.value = del_node.value
            if del_node == del_node_parent:
              current_node.right = del_node.right
              found_element = True
            if del_node.right == None:
              del_node_parent.left = None
              return
            else:
              del_node_parent.left = del_node.right
              return
  def traverseInorder(self, root):
    # Left -> Root -> Right
    res = []
    if root:
      res = self.traverseInorder(root.left)
      res.append(root.value) 
      res = res + self.traverseInorder(root.right)
    return res

  def printInorder(self, root): 
  
    if root: 
  
        # First recur on left child 
        self.printInorder(root.left) 
  
        # then print the data of node 
        print(root.value), 
  
        # now recur on right child 
        self.printInorder(root.right) 
```

**Binary Heaps** (not real relationship with the **heap memory**)

*  Every node at the top level has a greater value(max heap) or lower value(min heap) than the next level
*  Important when ordering is important
*  E.g Priority queues
*  Usually represented as arrays

Complexities

*   Lookup: O(n) Less efficient than BST, as it is less ordered
*   Insert: O(logn)
*   Delete: O(logn)

*Why use Binary Heap at the first place?*
Because they are good at comperative operations! Used in data storage, sorting algorithms

They is no ordering on the same level! If we swap the nodes on the same level it is still a binary heap. They do left to right insertions they preserve their structure => NOT UNBALANCED!

Pros:

1.   Priority (see priority heaps on udemy section)
2.   Better than O(n)
3.   Good at find max min O(1)

Cons:

1.   Slow lookup


**Implementation of Max heap**

```
import sys

class MaxHeap:
  # maxHeap constructor to create a heap with size = maxsize,
  # indixing starts at 1
  def __init__(self, maxSize):  # 
    self.maxSize = maxSize
    self.size = 0
    self.heap = [0] * (self.maxSize + 1) # the heap[0] will not be used for simplicity of indexing of parent and children nodes
    self.heap[0] = sys.maxsize # we set this as the max size the max value of int in order to perform comparisons

  # return the parent position of node at position idx
  def parent_index(self, node_idx):
    return node_idx // 2

  # The following functions in case the node has not children will not crash 
  # since the heap object is a list with static size
  # return the position of the left child of a node
  def left_child(self, node_idx):
    return 2 * node_idx
  
  # return the position of the right child of a node
  def right_child(self, node_idx):
    return 2 * node_idx + 1

  # returns true if a node is a leaf,
  # from theory in binary trees, we know that the number of the nodes in the last level
  # is equal the number of the nodes in the previous levels
  def is_leaf(self, node_idx):
    
    if node_idx >= (self.size // 2) and node_idx <= self.size:
      return True
    else:
      return False

  # swap the values of the nodes
  def swap(self, pos1, pos2):
    self.heap[pos1], self.heap[pos2] = self.heap[pos2], self.heap[pos1]


 # function to heapify the node inserted. This function is called to establish 
 # that the heap rule is valid
  def maxHeapify(self, node_idx):
    # check only for the indermediate nodes, if it is a leaf it will return
    if not self.is_leaf(node_idx):
      if self.heap[node_idx] < self.heap[self.left_child(node_idx)] or \
      self.heap[node_idx] < self.heap[self.right_child(node_idx)]:
        if self.heap[self.left_child(node_idx)] > self.heap[self.right_child(node_idx)]:
          self.swap(node_idx, left_child(node_idx)) # swap the node with its child
          self.maxHeapify(self, left_child(node_idx)) # run recursively for the child now as the node
        else: 
          self.swap(node_idx, right_child(node_idx)) # swap the node with its child
          self.maxHeapify(self, right_child(node_idx))

 # insert a node to the heap, we insert at the end and we swap
  def insert(self, item):
   if self.size == self.maxSize:
     print("The heap is Full")
     return
   else:
     self.size += 1 # increase the size of the heap
     self.heap[self.size] = item # add an item at the end
     current_node = self.size
     while self.heap[current_node] > self.heap[self.parent_index(current_node)]:
       #if self.heap[current_node] > self.heap[self.parent_index(current_node)]:
       print("It entered")
       self.swap(current_node, self.parent_index(current_node))
       current_node = self.parent_index(current_node) 


  #Method to print the contents of the heap in a detailed format
  def print_heap(self):
    for i in range(1, (self.size // 2) + 1):
      print(" PARENT : " + str(self.heap[i]) + " LEFT CHILD : " +
        str(self.heap[2 * i]) + " RIGHT CHILD : " +
        str(self.heap[2 * i + 1]))
      
  
```
